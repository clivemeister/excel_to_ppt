```python
from datetime import datetime
aug_df['Visit Date'].loc[1:5].apply(datetime.strptime,args=("%b %d, %Y",))
```

```python
def make_date(cell_val):
    if type(cell_val) is datetime:
        v=cell_val
    elif type(cell_val) is str: 
        try:
           v= datetime.strptime(cell_val,"%b %d, %Y").date()
        except ValueError:
            v=datetime.fromordinal(1).date
    else:
        v=datetime.fromordinal(2).date()
    return v

aug_df['Visit Date'].apply(make_date)
```

```python
all_df = pd.read_excel(open('Insights.xlsx','rb'),header=8)
all_df.head()
```

```python

def make_date(cell_val):
    if type(cell_val) is datetime:
        v=cell_val.date()
    elif type(cell_val) is str: 
        try:
            v= datetime.strptime(cell_val,"%b %d, %Y").date()
        except ValueError:
            v=datetime.fromordinal(1).date()
    else:
        v=datetime.fromordinal(2).date()
    return v

del all_df['date']

all_df.insert(loc=0,column='date',value=all_df['Visit Date'].apply(make_date))

from datetime import date
mask = (all_df['date']>=date(2017,8,1)) & (all_df['date']<=date(2017,8,31))
all_df.loc[mask].tail()
```
